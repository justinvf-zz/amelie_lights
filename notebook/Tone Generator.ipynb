```python
import numpy as np
from IPython.display import (Audio, display, clear_output)
from IPython.html import widgets
from __future__ import division
def play(hz):
    seconds = 10
    framerate = 44100
    clear_output()
    t = np.linspace(0, seconds, framerate*seconds)
    data = np.sin(2*np.pi*hz*t)
    display(Audio(data=data, autoplay=True, rate=framerate))
```

```python
hz_slider = widgets.IntSlider(min=40, max=10000)
w = widgets.interact(play, hz=hz_slider)
display(w)
```

```python
MIDDLE_A = 440
NOTES = ('A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#')
MIDDLE_KEYS = [(note, MIDDLE_A * 2** (i/12)) for (i, note) in enumerate(NOTES)]
ALL_KEYS = [('{}:{}'.format(i, note), freq * (2**i)) for i in range(-3, 4) for (note, freq) in MIDDLE_KEYS]
```

```python
def play_note(note):
    note, freq = ALL_KEYS[note]
    play(freq)
    print(note)
```

```python
note_slider = widgets.IntSlider(min=0, max=(len(ALL_KEYS) - 1))
w = widgets.interact(play_note, note=note_slider)
display(w)
```
